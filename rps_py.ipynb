```python
import random

def player(prev_play, opponent_history=[]):

    # Store opponent move
    if prev_play != "":
        opponent_history.append(prev_play)

    # If not enough data yet → play randomly
    if len(opponent_history) < 5:
        return random.choice(["R", "P", "S"])

    # Count opponent moves
    move_counts = {
        "R": opponent_history.count("R"),
        "P": opponent_history.count("P"),
        "S": opponent_history.count("S")
    }

    # Find most frequent opponent move
    most_played = max(move_counts, key=move_counts.get)

    # Counter strategy
    counter = {
        "R": "P",
        "P": "S",
        "S": "R"
    }

    return counter[most_played]
```
